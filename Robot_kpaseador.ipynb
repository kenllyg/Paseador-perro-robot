```python
import time
import random

class RobotPaseador:
    def __init__(self, nombre):
        self.nombre = nombre
        self.sensores = {"GPS": True, "Microfono": True, "Camara": True}
        self.inventario = []
        self.tiempo_paseo = 10  # Simulamos 10 segundos de paseo
        self.en_casa = True

    def check_sensores(self):
        print(f"--- PASO 1: Iniciando Checklist de {self.nombre} ---")
        for sensor, estado in self.sensores.items():
            if estado:
                print(f"[OK] {sensor} funcionando.")
            else:
                print(f"[ERROR] {sensor} falló.")
                return False
        print("Planificando ruta al parque más cercano...")
        return True

    def preparar_salida(self):
        print(f"\n--- PASO 2: Preparando salida ---")
        elementos = ["bolsas", "botella de agua", "correa"]
        for item in elementos:
            self.inventario.append(item)
            print(f"Cogiendo {item}...")

        print("Poniéndole la correa al perro. ¡Listo para salir!")
        self.en_casa = False

    def iniciar_paseo(self):
        print(f"\n--- PASO 3: Iniciando Paseo (Bucle de tiempo) ---")
        inicio = time.time()

        while time.time() - inicio < self.tiempo_paseo:
            tiempo_transcurrido = int(time.time() - inicio)
            print(f"Caminando... (Segundo {tiempo_transcurrido})")

            # Simulamos eventos aleatorios del entorno
            evento = random.choice(["nada", "charco", "caca", "pis", "nada"])

            if evento == "charco":
                print(">> [!] Charco detectado. Cambiando de acera.")

            elif evento == "caca":
                print(">> [!] El perro hizo caca. Recogiendo con bolsita y tirando a papelera.")

            elif evento == "pis":
                print(">> [!] El perro hizo pis en la pared. Echando agua de la botella.")

            if tiempo_transcurrido >= self.tiempo_paseo - 3:
                print(">> [AVISO] Queda poco tiempo. Girando hacia casa...")

            time.sleep(1) # Pausa de 1 segundo para simular tiempo real

    def finalizar_mision(self):
        print(f"\n--- PASO 4: Regreso y Limpieza ---")
        self.en_casa = True
        print("Entrando en casa y quitando correa.")

        patas = ["delantera derecha", "delantera izquierda", "trasera derecha", "trasera izquierda"]
        for pata in patas:
            print(f"Limpiando pata {pata} con toallita...")

        print("\nMisión completada. Robot en modo carga.")

# --- EJECUCIÓN DEL PROGRAMA ---
mi_robot = RobotPaseador("RoboDog-Walker v1.0")

if mi_robot.check_sensores():
    mi_robot.preparar_salida()
    mi_robot.iniciar_paseo()
    mi_robot.finalizar_mision()
else:
    print("Misión abortada por fallo técnico.")
```

Robot interactivo 2.0

```python
import time
import random
from datetime import datetime, timedelta

class RobotPaseadorPro:
    def __init__(self):
        self.nombre = "RoboWalker v3.0"
        self.sensores = {"GPS": True, "Cámara": True, "Micrófono": True}
        self.en_casa = True

    def menu_interactivo(self):
        print(f"\n{'='*40}")
        print(f"   🤖 SISTEMA DE PASEO: {self.nombre}   ")
        print(f"{'='*40}\n")

        # 1. Verificación de Sensores
        print("🔍 Iniciando autodiagnóstico...")
        time.sleep(1)
        if all(self.sensores.values()):
            print("✅ Todos los sistemas operativos.\n")
        else:
            print("❌ Error en sensores. Contacte a soporte.")
            return

        # 2. Interacción de Clima y Decisión
        clima_actual = random.choice(["despejado", "lluvioso", "nublado"])
        print(f"🌤️  Informe del tiempo: El cielo está {clima_actual.upper()}.")

        if clima_actual == "lluvioso":
            opcion = input("⚠️  ¡Atención! Está lloviendo. ¿Deseas sacar al perro de todas formas? (S/N): ").upper()
            if opcion != 'S':
                print("\n🏠 Decisión inteligente. Nos quedamos en casa jugando con la pelota.")
                return

        # 3. Elección de Ruta con validación
        print("\n📍 RUTAS DISPONIBLES:")
        print("1. Parque Central (Verde y tranquilo)")
        print("2. Paseo del Río (Fresco y con charcos)")
        print("3. Avenida Principal (Rápido y urbano)")

        while True:
            try:
                seleccion = int(input("\nSelecciona el número de la ruta deseada: "))
                rutas = {1: "Parque Central", 2: "Paseo del Río", 3: "Avenida Principal"}
                if seleccion in rutas:
                    self.ruta = rutas[seleccion]
                    break
                else:
                    print("❌ Por favor, elige 1, 2 o 3.")
            except ValueError:
                print("❌ Entrada no válida. Introduce un número.")

        # 4. Definición de tiempo
        while True:
            try:
                self.duracion = int(input(f"⏱️  ¿Cuántos minutos durará el paseo por {self.ruta}?: "))
                if self.duracion > 0:
                    break
                print("❌ El tiempo debe ser mayor a 0.")
            except ValueError:
                print("❌ Por favor, introduce un número entero.")

        # Iniciar misión
        self.ejecutar_paseo()

    def ejecutar_paseo(self):
        hora_inicio = datetime.now()
        hora_fin = hora_inicio + timedelta(minutes=self.duracion)

        print(f"\n{'*'*40}")
        print(f"🚀 INICIANDO PASEO")
        print(f"⏰ Hora de salida: {hora_inicio.strftime('%H:%M:%S')}")
        print(f"🏁 Hora prevista de regreso: {hora_fin.strftime('%H:%M:%S')}")
        print(f"{'*'*40}\n")

        # Simulación del paso del tiempo (acelerada para el ejemplo)
        # En un caso real usaríamos la duración ingresada
        pasos_simulados = 5
        for i in range(pasos_simulados):
            print(f"🚶 Caminando por {self.ruta}... [{i+1}/{pasos_simulados}]")

            # Eventos durante el paseo
            if random.random() < 0.3: # 30% de probabilidad de evento
                evento = random.choice([
                    "💩 El perro hizo sus necesidades. Usando bolsa...",
                    "💦 El perro marcó una pared. Limpiando con agua...",
                    "🌊 ¡Charco a la vista! Desviando ruta..."
                ])
                print(f"   ✨ {evento}")

            time.sleep(1.5)

        self.finalizar()

    def finalizar(self):
        print(f"\n{'='*40}")
        print("🏡 LLEGADA A CASA")
        print(f"⏰ Hora actual: {datetime.now().strftime('%H:%M:%S')}")
        print("🧼 Limpiando las 4 patitas con toallitas húmedas...")
        time.sleep(1)
        print("🐶 ¡Perro feliz y limpio! Misión terminada.")
        print(f"{'='*40}")

# --- EJECUCIÓN ---
robot = RobotPaseadorPro()
robot.menu_interactivo()
```

```python
inventario = 100
print(f"El inventario contiene {inventario} hamburguesas")
while inventario > 0:
  if inventario <= 10:
    print("ADVERTENCIA: EL INVENTARIO ESTA CASI VACÍO")
  num_hamburguesas = int(input("Cuántas hamburguesas quiere comprar? "))
  if num_hamburguesas > inventario:
      print("No hay suficientes hamburguesas, solo quedan", inventario, "disponibles")
  else:
      inventario -= num_hamburguesas
      print(f"Haz pedido", num_hamburguesas, "Sólo nos quedan", inventario, "hamburguesas")
```
